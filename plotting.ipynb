# Plotting

```julia
using Pkg

Pkg.add("Plots")
using Plots
```

```julia
x = -3:0.1:3
f(x) = x ^ 2

y = f.(x)
```

```julia
# GR backend

gr()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")

# obs, without the ! after scatter, the line plot would be changed
```

```julia
Pkg.add("PlotlyJS")
```

```julia
plotlyjs()
plot(x, y, label="line")
scatter!(x, y, label="points")
```

# Getting slightly fancier

```julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17];
```

```julia
plot(numpirates, globaltemperatures, legend=false)
scatter!(numpirates, globaltemperatures, legend=false)

xflip!()

xlabel!("Number of pirates [Approximate]")
ylabel!("Global temperature (C)")
title!("Influence of pirate population on global temperature")
```

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1, p2, p3, p4, layout=(2, 2), legend=false)
```
